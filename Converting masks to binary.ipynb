```python
#Converting the ground truth masks to binary
import cv2
import os
import imageio
path='data/Masks/'
path1='data/binary/'
for i in range(199):
    img = cv2. imread(path + str(i) + '_expert.bmp',2)
    ret, bw_img = cv2. threshold(img,0,255,cv2. THRESH_BINARY)
    #cv2. imshow("Binary Image",bw_img)
    imageio.imwrite(path1 + str(i) + '.png',bw_img)
    #cv2. waitKey(0)
    #cv2. destroyAllWindows()
```
